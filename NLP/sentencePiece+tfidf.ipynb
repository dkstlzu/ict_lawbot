```python
f=open('total_cases.txt', 'r', encoding='euc-kr')

data=[]
lines=f.readlines()

# with open("total_cases.txt", "rb") as input:
#     data = input.read()

# # 바이너리 데이터를 utf-16으로 디코딩한다
# data = data.decode("utf-16")

# # 수직 탭을 삭제한다
# data = data.replace(u"\u000B", u"")

# # 유니코드 데이터를 utf-8로 인코딩한다
# data = data.encode("utf-8")

for line in lines :
  data.append(line)

data
```

```python
data_train=[]
for sentence in data:
    temp_X = src_sp.EncodeAsPieces(sentence)

    # temp_X=okt.morphs(sentence, stem=True) # 토큰화
    temp_X=[word for word in temp_X if not word in stopwords] # 불용어 제거
    data_train.append(temp_X)

data_train
```

```python
realdata=[]

for sentence in data_train :
  test=' '.join(sentence).replace('▁', '')
  realdata.append(test)
  

realdata
```

```python
tfidfv=TfidfVectorizer().fit(realdata)
print(tfidfv.transform([test]).toarray())
print(tfidfv.vocabulary_)
```

```python
vocablist=tfidfv.transform([test]).toarray()

idx=0
for a in vocablist[0] :
  if a!=0 :
    print(str(idx)+" "+str(a))
  idx+=1
# vocablist
```
